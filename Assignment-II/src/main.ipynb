```python
import numpy as np
import matplotlib.pyplot as plt

# System values (from your STREAM + compute)
peak_compute = 36  # GFLOP/s
measured_bandwidth = 23.72  # GB/s
ridge_point = peak_compute / measured_bandwidth  # FLOPs/Byte

# Your measured point
measured_oi = 1 / 24  # ≈ 0.0417 FLOPs/Byte
measured_perf = 0.723  # GFLOP/s

# Range for the memory limit line
oi_range = np.linspace(0.01, 5, 200)
memory_roof = measured_bandwidth * oi_range

# Plotting
plt.figure(figsize=(8,6))
plt.loglog(oi_range, memory_roof, linestyle='dotted', color='gray', label=f'Memory Bandwidth Limit ({measured_bandwidth:.2f} GB/s)')
plt.axhline(y=peak_compute, color='red', linestyle='--', label=f'Compute Roof ({peak_compute} GFLOP/s)')
plt.plot(measured_oi, measured_perf, 'bo', markersize=10, label=f'Measured Point\n({measured_oi:.4f}, {measured_perf:.3f})')

# Labels and aesthetics
plt.xlabel('Operational Intensity (FLOPs/Byte)')
plt.ylabel('Performance (GFLOP/s)')
plt.title('Roofline Model (Single-Core, Dardel)')
plt.legend()
plt.grid(True, which='both', linestyle='--')
plt.tight_layout()
plt.show()
```
