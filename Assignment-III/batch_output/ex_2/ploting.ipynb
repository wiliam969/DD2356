```python
#didnt try to hassle with notebook so i used the code cells below in google colab

import re
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict
import glob
import os

# Gather all .stdout files
stdout_files = glob.glob("smp_output_*_run*.stdout")

# Regex patterns
filename_pattern = re.compile(r"smp_output_(\d+)_run\d+\.stdout")
bw_pattern = re.compile(r"Copy:\s+([\d.]+)")

# Data structure: {thread_size: [copy_bandwidth_values]}
copy_data = defaultdict(list)

for filepath in stdout_files:
    filename = os.path.basename(filepath)
    thread_match = filename_pattern.match(filename)
    if not thread_match:
        continue
    thread_size = int(thread_match.group(1))

    with open(filepath, 'r') as f:
        content = f.read()

    match = bw_pattern.search(content)
    if match:
        copy_data[thread_size].append(float(match.group(1)))

# Prepare plot data
thread_sizes = sorted(copy_data.keys())
means = [np.mean(copy_data[t]) for t in thread_sizes]
stds = [np.std(copy_data[t]) for t in thread_sizes]

# Plotting
plt.figure(figsize=(8, 5))
plt.errorbar(thread_sizes, means, yerr=stds, fmt='o-', color='blue',
             ecolor='black', capsize=5, markersize=8, label='Copy Bandwidth')

plt.title("STREAM Copy Bandwidth vs Thread Count")
plt.xlabel("Thread Count")
plt.ylabel("Bandwidth (MB/s)")
plt.xticks(thread_sizes)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import re
import matplotlib.pyplot as plt
import numpy as np
from collections import defaultdict
import glob
import os

# Gather all .stdout files for 128-thread runs
stdout_files = glob.glob("smp_output_128_*_run*.stdout")

# Regex patterns
filename_pattern = re.compile(r"smp_output_128_(\w+)_run\d+\.stdout")
bw_pattern = re.compile(r"Copy:\s+([\d.]+)")

# Data structure: {schedule_type: [copy_bandwidth_values]}
copy_data = defaultdict(list)

for filepath in stdout_files:
    filename = os.path.basename(filepath)
    match = filename_pattern.match(filename)
    if not match:
        continue
    schedule_type = match.group(1)

    with open(filepath, 'r') as f:
        content = f.read()

    bw_match = bw_pattern.search(content)
    if bw_match:
        copy_data[schedule_type].append(float(bw_match.group(1)))

# Prepare plot data
schedule_types = sorted(copy_data.keys())
means = [np.mean(copy_data[sched]) for sched in schedule_types]
stds = [np.std(copy_data[sched]) for sched in schedule_types]

# Plotting
plt.figure(figsize=(8, 5))
plt.errorbar(schedule_types, means, yerr=stds, fmt='o-', color='green',
             ecolor='black', capsize=5, markersize=8, label='Copy Bandwidth')

plt.title("STREAM Copy Bandwidth for Scheduling Types (128 Threads)")
plt.xlabel("Schedule Type")
plt.ylabel("Bandwidth (MB/s)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
